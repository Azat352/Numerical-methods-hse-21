```python
import numpy as np
import matplotlib.pyplot as plt
```

# Задание 4

```python
def implicit_euler_solve(u0, T, dt):
    """Решает $du/dt = \frac{1}{u}$ на $0 < t < T$ с $u(t=0) = u0$ при помощи неявного метода Эйлера."""
    num_steps = int(T/dt)
    tt = np.arange(num_steps+1)*dt
    y_p = np.empty(num_steps + 1)
    y_p[0] = u0
    
    for i in range(num_steps):
        y_p[i + 1] = (y_p[i] + np.sqrt(y_p[i]**2 + 4*dt))/2
    return tt, y_p
```

```python
def true_result(t):
    return np.sqrt(2) * np.sqrt(t + 1/2)
```

```python
tt, y_p = implicit_euler_solve(1, T=5, dt=0.1)
tt_t = np.linspace(0, 5, 1000)
plt.plot(tt, y_p, color='blue', label='euler')

plt.plot(tt_t, true_result(tt_t), color='red', label='true function')

plt.legend(loc='best')
plt.grid()
```
